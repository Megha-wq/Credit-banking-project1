```python
#Impose an interest rate of 2.9% for each customer for any due amount.
# Task8
def impose_interest(data_final):
  """
  This function imposes an interest rate of 2.9% for each customer for any due amount.

  Args:
    data_final: A Pandas DataFrame containing the customer data.

  Returns:
    A Pandas DataFrame with the updated due amounts after interest is applied.
  """

  # Calculate the interest amount for each customer
  data_final["Interest"] = data_final["Limit"] * 2.9 / 100

  # Update the due amount with interest
  data_final["Limit"] = data_final["Limit"] + data_final["Interest"]

  return data_final

# Apply the function to the data_final DataFrame
data_final = impose_interest(data_final)

# Print the updated DataFrame
print(data_final)
```

```python
# Monthly profit for the bank
#Task9
monthly_profit = data_final["Limit"].sum() * 0.15
print(f"The monthly profit for the bank is: {monthly_profit}")
```
